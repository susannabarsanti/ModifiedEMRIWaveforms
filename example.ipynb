```python
import few 
from few.trajectory.inspiral import EMRIInspiral
from few.waveform import GenerateEMRIWaveform
from few.trajectory.ode import KerrEccEqFlux
from few.utils.geodesic import get_separatrix

```

```python
from Scalar_Trajectories import KerrCircEqFluxScalar
```

```python
M = 1e6
mu = 10
a = 0.9
e0 = 0.0
x_I0 = 1.0
dt = 1.0
Tobs = 1 

print("M = ", M)
print("mu = ", mu)
print("a = ", a)
print("e0 = ", e0)
print("x_I0 = ", x_I0)
print("dt = ", dt)
print("Tobs = ", Tobs)
```

```python
p_sep = get_separatrix(a, e0, x_I0)
print(f"Separatrix: {p_sep}")
```

```python
p_buffer = 0.01
```

```python
traj_gr = EMRIInspiral(func=KerrEccEqFlux)
t_back, p_back, e_back, x_back, Phi_phi_back, Phi_r_back, Phi_theta_back = traj_gr(M, mu, a, p_sep + p_buffer, e0, x_I0 , dt=dt, T=Tobs, integrate_backwards=True)

p0 = p_back[-1]

print("p0 = ", p0)
```

```python
d = 0.0
trajKerrScalar = EMRIInspiral(func=KerrCircEqFluxScalar, integrate_constants_of_motion=False)
t_back, p_back, e_back, x_back, Phi_phi_back, Phi_r_back, Phi_theta_back = trajKerrScalar(M, mu, a, p_sep + p_buffer, e0, x_I0 , d,  dt=dt, T=Tobs, integrate_backwards=True)

p0 = p_back[-1]
print("p0 set to ", p0, flush=True)
```
